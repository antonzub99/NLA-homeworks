# 1.1

_1._
Let's recall what are L1-norm and Chebyshev norm:
$$
||x||_1 = \sum\limits_{k=1}^n |x_k|\\
||x||_{\infty} = \underset{1\leq k \leq n}{\max} |x_k|
$$
Then we can bound L1-norm as $||x||_1 \leq \sum\limits_{k=1}^n \underset{1\leq k \leq n}{\max} |x_k| = n ||x||_{\infty}$. 

Thus, 
$$
\dfrac{1}{n}||x||_1 \leq ||x||_{\infty}
$$
On the other hand $||x||_1 = \sum\limits_{k=1}^n |x_k| = ||x||_{\infty}\sum\limits_{k=1}^n \dfrac{|x_k|}{||x||_{\infty}}$. In this sum at least one element equals to 1, and other are less then 1. 

Thus, $||x||_1 \geq ||x||_{\infty}$.
Combining this inequalities we obtain

$$
\dfrac{1}{n}||x||_1 \leq ||x||_{\infty} \leq ||x||_1
$$
So, $C_1 = \dfrac{1}{n}$ and $C_2 = 1$

_2._
There is a convinient form for Frobenius norm $||A||_F^2 = tr\left(AA^*\right)$, where $A^*$ is hermitian conjugate of $A$.
Unitary matrix $U$ is such a matrix, that $UU^*=U^*U = I$. 

Recall that $tr\left(ABC\right)=tr\left(BCA\right)=tr\left(CAB\right)$. 

Then
$$
||UA||_F^2 = tr\left(UA\left(UA\right)^*\right) = tr\left(UAA^*U^*\right) = tr\left(AA^*U^*U\right) = tr\left(AA^*I\right) = tr\left(AA^*\right) = ||A||_F^2\\
||AU||_F^2 = tr\left(AUU^*A^*\right) = tr\left(AIA^*\right) = tr\left(IA^*A\right) = tr\left(A^*A\right)=||A||_F^2
$$

_3._
Given linear metric spaces $V$ and $W$ with corresponding scalar products $(\cdot, \cdot)_V$, $(\cdot, \cdot)_W$ and operator $U: V \rightarrow W$, $U$ is called unitary if $\forall x,y \in V$: $(Ux, Uy)_W = (x,y)_V$. Thus, $(Ux, Ux)_W = (x,x)_V$ and $U$ preserves the corresponding norm, induced by scalar product. L2-norm for vectors is induced by standard scalar product, so unitary operators (and thus unitary matrices) preserve L2-norm: $||Ux||_2 = ||x||_2$ 

Spectral norm $||A||_2$ is defined as $||A||_2 = \underset{||x||_2=1}{\sup} ||Ax||_2$, so it's connected with L2-norm of vectors. Then we immediately obtain
$$
||UAx||_2 = ||Ax||_2 \rightarrow ||UA||_2 = \underset{||x||_2=1}{\sup} ||UAx||_2 = \underset{||x||_2=1}{\sup} ||Ax||_2 = ||A||_2\\
||AUx||_2 = ||AUU^*x||_2 = ||AIx||_2 = ||Ax||_2 \rightarrow ||AU||_2 = ||A||_2
$$

_1._
Let's use the trace form of Frobenius norm once again and combine it with the SVD decomposition of the matrix $A: A = V\Sigma U^*$. Recall, that $V, U$ are unitary matrices, and $\Sigma = diag\left(\sigma_1,...,\sigma_r\right), r=rank \left(A\right)$ (for rectangular matrices $A$ $\Sigma$ will be also rectangular, and entries of $\Sigma$ are always real, then $\Sigma^* = \Sigma^T$)
$$
||A||_F^2 = tr\left(AA^*\right) = tr\left(V\Sigma U^* \left(V\Sigma U^*\right)^*\right) = tr\left(V\Sigma U^* U \Sigma^* V^*\right) = tr\left(I\Sigma^*I\Sigma \right) = tr\left(\Sigma \Sigma^*\right) = \sum\limits_{k=1}^r \sigma_k^2 \leq r \sigma_1^2
$$
It can be shown, that $||A||_2 = \underset{||x||_2=1}{\sup} ||Ax||_2 = \sigma_1$, then 
$$
||A||_F \leq \sqrt{r} \sigma_1 = \sqrt{r} ||A||_2 = \sqrt{rank\left(A\right)} ||A||_2
$$

_2._
Directly from previous exercise it's seen, that equality is satisfied iff all the singular values of matrix $A$ are the same: $\sigma_i = \sigma_1, i=\overline{1,r}$.

_3._
Diagonalizable matrix $A \in \mathbb{R}^{n \times n}$ can be decomposed as $A = S^{-1}\Lambda S$, where $\Lambda = diag\left(\lambda_1,...,\lambda_n\right)$ is matrix of eigenvalues of $A$, and $S$ is a matrix of eigenvectors of $A$.

By definition of matrix exponent $\exp{(A)} = \sum\limits_{k=0}^{\infty}\dfrac{A^k}{k!}$, and for diagonal matrix $\Lambda$: $\exp{(\Lambda)} = diag\left(e^{\lambda_1},...,e^{\lambda_n}\right)$.

Then
$$
\det{(\exp{(A)})} = \det{(\exp{(S^{-1}\Lambda S)})} = \det{\left(\sum\limits_{k=1}^{\infty}\dfrac{\left(S^{-1}\Lambda S\right)^k}{k!}\right)} = \det{(S^{-1}\exp{(\Lambda)}S)} = \\
=\det{(S^{-1})} \cdot \det{(diag\left(e^{\lambda_1},..., e^{\lambda_n}\right))} \cdot \det{(S)}
$$
Recall, that $\det{(S^{-1})} = \dfrac{1}{\det{(S)}}$, $\det{(diag(\alpha_k))} = \prod\limits_{k=1}^n \alpha_k$ and $tr\left(A\right) = tr\left(S^{-1}\Lambda S\right) = tr\left(\Lambda SS^{-1}\right)=tr\left(\Lambda\right)$.
$$
\det{(\exp{(A)})} = \det{(diag\left(e^{\lambda_1},..., e^{\lambda_n}\right))} = \prod\limits_{k=1}^n e^{\lambda_k} = e^{\sum\limits_{k=1}^n \lambda_k} = \exp{(tr\left(\Lambda\right))} = \exp{(tr\left(A\right))}
$$

Thus, $tr\left(B^*B\right) = tr\left(diag\left(\lambda_1,...,\lambda_r \right)\right) = \sum\limits_{k=1}^r \lambda_k = \sum\limits_{k=1}^r \sigma_k^2$
$$
||B||_2^2 = (\underset{||x||_2=1}{\sup} ||Bx||_2)^2 = \sigma_1^2 \leq \sum\limits_{k=1}^r \sigma_k^2 = tr\left(B^* B\right) = ||B||_F^2\\
||B||_2 \leq ||B||_F
$$

Assembling inequalities

$$

\lambda^2 - \lambda = 0 \rightarrow \lambda = 0, 1
$$
Now, for SVD of $UU^* = \hat{U} \Sigma V^*, \Sigma$ consists of square roots of eigenvalues of matrix $(UU^*)^* UU^* = UU^* UU^* = UU^*$, but we've already noted that it's a projector, so eigenvalues of $(UU^*)^* UU^*$ are as the ones of any projector: $0$ and $1$.

Thus, $\Sigma$ contains only ones and zeros and it's inverse $\Sigma^{\dagger}$ also contains only ones and zeros. We can formally write $\Sigma^{\dagger} = \Sigma^{*}$.

$$
(UU^*)^{\dagger} = V \Sigma^{\dagger} U^* = V \Sigma^* U^* = (U \Sigma V^*)^* = (UU^*)^* = UU^*
$$
So pseudoinverse of $UU^*$ is itself.

# 2.1

Firstly, let's rewrite $F(U,V):$
$$
F(U,V) = \dfrac{1}{2}||X-UDV||_F^2 = \dfrac{1}{2}tr\left((X-UDV)(X-UDV)^T\right) = \dfrac{1}{2}tr\left((X-UDV)(X^T-V^T D^T U^T)\right), D=D^T\\
F(U,V) = \dfrac{1}{2} tr\left(XX^T - X V^T D U^T - UDVX^T + UDVV^TDU^T\right) = \dfrac{1}{2}\left[tr(XX^T) - tr(X V^T D U^T) - tr(UDVX^T) + tr(UDVV^TDU^T)\right]
$$
Secondly, we should derive some basic formulas of derivatives of traces w.r.t matrices. Let's use Einstein's rule of summation by repeated indices for simplicity
$$
\left(\dfrac{\partial \mathbf{X}}{\partial \mathbf{X}}\right)_{ijlm} = \dfrac{\partial x_{ij}}{\partial x_{lm}} = \delta_{il} \delta_{jm} - \text{general formula}, \delta_{il} = \begin{cases}1, i=l\\0, i \ne l \end{cases} - \text{Kronecker delta}\\
\left(\dfrac{\partial tr(\mathbf{AX})}{\partial \mathbf{X}}\right)_{ij} = \dfrac{\partial a_{nm}x_{mn}}{\partial x_{ij}} = a_{nm} \delta_{mi} \delta_{nj} = a_{ji} = (A^T)_{ij}\\
\left(\dfrac{\partial tr(\mathbf{AX^T})}{\partial \mathbf{X}}\right)_{ij} = \dfrac{\partial a_{nm}x_{nm}}{\partial x_{ij}} = a_{nm} \delta_{ni} \delta_{mj} = a_{ij} = (A)_{ij}\\
\left(\dfrac{\partial tr(\mathbf{X^TAX})}{\partial \mathbf{X}}\right)_{ij} = \dfrac{\partial x_{mn}a_{ml}x_{ln}}{\partial x_{ij}} = \delta_{mi} \delta_{nj} a_{ml} x_{ln} + x_{mn}a_{ml} \delta_{li} \delta_{nj} = a_{il} x_{lj} + x_{mj}a_{mi} = (AX)_{ij} + (A^T X)_{ij} = ((A+A^T)X)_{ij}\\
\left(\dfrac{\partial tr(\mathbf{XAX^T})}{\partial \mathbf{X}}\right)_{ij} = \dfrac{\partial x_{nm}a_{ml}x_{nl}}{\partial x_{ij}} = \delta_{ni} \delta_{mj} a_{ml} x_{nl} + x_{nm}a_{ml} \delta_{ni} \delta_{lj} = a_{jl} x_{il} + x_{im}a_{mj} = (XA^T)_{ij} + (XA)_{ij} = (X(A^T+A))_{ij}
$$


_1._
$$
\nabla_U F(U,V) = \dfrac{1}{2}\dfrac{\partial}{\partial \mathbf{U}}\left(tr(XX^T) - tr(X V^T D U^T) - tr(UDVX^T) + tr(UDVV^TDU^T)\right) = \dfrac{1}{2}\left(-XV^TD-XV^TD + (1)\right)\\
(1) = \dfrac{\partial(tr(UDVV^TDU^T))}{\partial \mathbf{U}} = U((DVV^TD)^T + DVV^TD) = U(DVV^TD + DVV^TD) = 2UDVV^TD\\
\nabla_U F(U,V) = UDVV^TD - XV^TD = (UDV - X)V^T D
$$

_2._
$$
\nabla_V F(U,V) = \dfrac{1}{2}\dfrac{\partial}{\partial \mathbf{V}}\left(tr(XX^T) - tr(X V^T D U^T) - tr(UDVX^T) + tr(UDVV^TDU^T)\right)=\\
=\dfrac{1}{2}\dfrac{\partial}{\partial \mathbf{V}}\left(tr(XX^T) - tr(D U^T X V^T) - tr(X^TUDV) + tr(V^TDU^TUDV)\right) = DU^TUDV - DU^TX = DU^T (UDV - X)
$$

_3._
We can see that evaluating the $\nabla_{U} F(U,V)$ takes 
$$
U_{n \times k} D_{k \times k} - O(nk^2)\\ 
(UD)_{n \times k} V_{k \times n} - O(n^2 k)\\ 
(UDV)_{n \times n} - X_{n \times n} - O(n^2)\\
(UDV - X)_{n \times n} V^T_{n \times k} - O(n^2 k)\\ 
((UDV - X)V^T)_{n \times k} D_{k \times k} - O(nk^2) \rightarrow \\ 
2 O(nk^2) + 2O(n^2 k) + O(n^2) = O(nk(n+k))
$$
For $\nabla_{V} F(U,V)$
$$
(UDV - X) - O(nk^2 + n^2k + n^2)\\ 
D_{k \times k} U^T_{k \times n} - O(k^2n)\\ 
(DU^T)_{k \times n} (UDV - X)_{n \times n} - O(kn^2) \rightarrow \\ 
O(nk(n+k))
$$
Which was an obvious result, as it takes the same amount of multiplcation of the same sized matrices. Now let's compare the bruteforce analytical calculation with jax calculations.


````python
import jax
import jax.numpy as jnp
from timeit import timeit

@jax.jit
def func_F(X, U, V, D):
    return 0.5 * jnp.square(jnp.linalg.norm(X - U @ D @ V))

def brute_gradU(X, U, V, D):
    return (U @ D @ V - X) @ V.T @ D

def brute_gradV(X, U, V, D):
    return D @ U.T @ (U @ D @ V - X) 

gradU = jax.jit(jax.grad(func_F, argnums=1))
gradV = jax.jit(jax.grad(func_F, argnums=2))
````

````python
from tqdm.notebook import tqdm
rows = jnp.array([50, 100, 250, 500, 750, 1000, 2000, 5000])
fixed = 100
keyX, keyU, keyV, keyD = jax.random.split(jax.random.PRNGKey(1337), 4)

rows_brute_U = []
rows_brute_V = []
for it in tqdm(rows):
    X = jax.random.normal(keyX, (it, it))
    U = jax.random.normal(keyU, (it, fixed))
    D = jnp.diag(jax.random.normal(keyD, (fixed, )))
    V = jax.random.normal(keyV, (fixed, it))

    timeU = %timeit -oq brute_gradU(X, U, V, D)
    rows_brute_U.append(timeU.best)
    
    timeV = %timeit -oq brute_gradV(X, U, V, D)
    rows_brute_V.append(timeV.best)
````

````python
rows_jax_U = []
rows_jax_V = []
for it in tqdm(rows):
    X = jax.random.normal(keyX, (it, it))
    U = jax.random.normal(keyU, (it, fixed))
    D = jnp.diag(jax.random.normal(keyD, (fixed, )))
    V = jax.random.normal(keyV, (fixed, it))

    timeU = %timeit -oq gradU(X, U, V, D).block_until_ready()
    rows_jax_U.append(timeU.best)
    
    timeV = %timeit -oq gradV(X, U, V, D).block_until_ready()
    rows_jax_V.append(timeV.best)
````

````python
cols = jnp.array([50, 100, 250, 500, 750, 1000, 2000, 5000])
fixed = 100
keyX, keyU, keyV, keyD = jax.random.split(jax.random.PRNGKey(1337), 4)

cols_brute_U = []
cols_brute_V = []
for it in tqdm(rows):
    X = jax.random.normal(keyX, (fixed, fixed))
    U = jax.random.normal(keyU, (fixed, it))
    D = jnp.diag(jax.random.normal(keyD, (it, )))
    V = jax.random.normal(keyV, (it, fixed))

    timeU = %timeit -oq brute_gradU(X, U, V, D)
    cols_brute_U.append(timeU.best)
    
    timeV = %timeit -oq brute_gradV(X, U, V, D)
    cols_brute_V.append(timeV.best)
````

````python
cols_jax_U = []
cols_jax_V = []
for it in tqdm(rows):
    X = jax.random.normal(keyX, (fixed, fixed))
    U = jax.random.normal(keyU, (fixed, it))
    D = jnp.diag(jax.random.normal(keyD, (it, )))
    V = jax.random.normal(keyV, (it, fixed))

    timeU = %timeit -oq gradU(X, U, V, D).block_until_ready()
    cols_jax_U.append(timeU.best)
    
    timeV = %timeit -oq gradV(X, U, V, D).block_until_ready()
    cols_jax_V.append(timeV.best)
````

$$
f(x) = \dfrac{(Ax,x)}{(x,x)} = \dfrac{x^TAx}{x^Tx} = \dfrac{a_{ij}x_i x_j}{x_p x_p}, a_{ij} = a_{ji}
$$
Let's compute first gradient (in fact, when writing down gradients in such a form the denominator index must be upper, as it's a one-form and transforms covariantly with the change of basis, but we don't care about this here)
$$
\nabla_x f(x) = \dfrac{\partial f(x)}{\partial \mathbf{x}} \rightarrow \left(\nabla_x f(x)\right)_i = \dfrac{\partial f(x)}{\partial x_i}=\\
=\dfrac{\partial}{\partial x_i} \dfrac{a_{mn}x_m x_n}{x_p x_p} = \dfrac{2 a_{im} x_m}{x_p x_p} - \dfrac{2 a_{mn} x_m x_n x_i}{(x_p x_p)^2} = \dfrac{2}{x^T x} (Ax)_i - \dfrac{2}{x^T x} f(x)x_i\\
\nabla_x f(x) = \dfrac{2}{x^T x} (A - \mathbb{1}f(x))x
$$

Now we compute Hessian
$$
\mathbf{H}_f = \dfrac{\partial f}{\partial \mathbf{x} \partial \mathbf{x}^T} \rightarrow (\mathbf{H}_f)_{ij} = \dfrac{\partial f}{\partial x_i \partial x_j} = \\
= \dfrac{\partial}{\partial x_j} \left(\dfrac{2 a_{il}x_l}{x_p x_p}\right) - \dfrac{\partial}{\partial x_j} \left(f(x) \dfrac{x_i}{x_p x_p}\right) = \dfrac{2 a_{ij} x_p x_p}{(x_p x_p)^2} - \dfrac{4x_j a_{il}x_l}{(x_p x_p)^2} - \dfrac{\partial f(x)}{\partial x_j}\dfrac{x_i}{x_p x_p} - f(x)\dfrac{\partial}{\partial x_j} \left(\dfrac{x_i}{x_p x_p}\right)=\\
= \dfrac{2}{x^T x} a_{ij} - \dfrac{4}{(x^Tx)^2} a_{il}x_l x_j - \\
- \dfrac{4}{(x^Tx)^2}a_{jl}x_l x_i + \dfrac{4}{(x^Tx)^2}\dfrac{x^TAx}{x^Tx} x_i x_j - \\
-\dfrac{2}{x^Tx}\dfrac{x^TAx}{x^Tx} \delta_{ij} + \dfrac{4}{(x^Tx)^2}\dfrac{x^TAx}{x^Tx}x_ix_j=\\
= \dfrac{2}{x^Tx}(a_{ij} - f(x)\delta_{ij}) + \dfrac{4}{(x^Tx)^2}\left(2f(x)x_i x_j - (Ax)_i x_j - (Ax)_j x_i\right) =\\
=\dfrac{2}{x^Tx}(a_{ij} - f(x)\delta_{ij}) - \dfrac{4}{(x^Tx)^2}\left(a_{il}x_l x_j + a_{jl}x_l x_i- 2f(x)x_i x_j \right)
$$
In a matrix form
$$
\mathbf{H}_f = \dfrac{2}{x^Tx}(A - \mathbb{1}f(x)) - \dfrac{4}{(x^Tx)^2}(Axx^T + xx^T A - 2f(x)xx^T) =\\
= \dfrac{2}{x^Tx}\left(A - \mathbb{1}f(x)\right) - \dfrac{2}{x^Tx} \nabla_x f(x) x^T - \dfrac{2}{x^Tx} x (\nabla_x f(x))^T = \\
= \dfrac{2}{x^T x} \left(A - \mathbb{1}f(x) - \left(\nabla_x f(x)\right) x^T - x (\nabla_x f(x))^T\right)
$$

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

Singular values starting from 200th to 800th are changing just a little (compared to the head and the tale of each plot). Thus, we can clip off the last 50-100 singular values and get the approximation of each matrix with a matrix of a lesser rank, but this approximation won't give us a big reduction of memory usage. So we should study the appropriate comprassion rank in the next subexercise.

````python
class CompressedFC(nn.Module):
    def __init__(self, svds, biases):
        super().__init__()
        self.U = svds[0]
        self.S = svds[1]
        self.VT = svds[2]
        self.bias = biases[0]
        
    def forward(self, x):
        x = self.VT @ x.T
        x = torch.diag(self.S) @ x
        x = self.U @ x
        x = torch.cat(x.shape[1] * [torch.unsqueeze(self.bias, 1)], 1) + x
        return x


class CompressedResBlock(nn.Module):
    def __init__(self, svds, biases, activation=nn.ReLU()):
        super().__init__()
        self.compressed_layer1 = svds[0]
        self.bias1 = biases[0]
        self.compressed_layer2 = svds[1]
        self.bias2 = biases[1]
        self.activation = activation

    def forward(self, x):
        out = x.T
        #out = USV^T x, so we traverse the list with matrices backwards
        for item in reversed(self.compressed_layer1):
            #as input is a batched vector, transposition is performed to keep outputs batched
            out = (item @ out.T).T
        out = torch.cat(out.shape[0] * [torch.unsqueeze(self.bias1, 0)], 0) + out
        out = self.activation(out)
        
        for item in reversed(self.compressed_layer2):
            out = (item @ out.T).T
        out = torch.cat(out.shape[0] * [torch.unsqueeze(self.bias2, 0)], 0) + out
        out = out.T + x

        out = self.activation(out)
        return out



class CompressedNet(nn.Module):
    def __init__(self, rank, Net, factor_grad=True):
        super(CompressedNet, self).__init__()
        self.rank = rank
        self.Net = Net
        self.U = nn.ParameterList([])
        self.S = nn.ParameterList([])
        self.VT = nn.ParameterList([])
        self.activation = nn.ReLU()

        for params in self.Net.parameters():
            if (len(params.shape)==2 and params.shape[0]>10):
                U, S, VT = torch.linalg.svd(params)
                self.U.append(nn.Parameter(U[:, :self.rank], requires_grad=factor_grad))
                self.S.append(nn.Parameter(S[:self.rank], requires_grad=factor_grad))
                self.VT.append(nn.Parameter(VT[:self.rank,:], requires_grad=factor_grad))    

        self.fc1 = CompressedFC([self.U[0], self.S[0], self.VT[0]], [self.Net.fc1.bias])
        self.res1 = CompressedResBlock([[self.U[1], torch.diag(self.S[1]), self.VT[1]], 
                                        [self.U[2], torch.diag(self.S[2]), self.VT[2]]],
                                       [self.Net.res1.layer1.bias, self.Net.res1.layer2.bias])
        self.res2 = CompressedResBlock([[self.U[3], torch.diag(self.S[3]), self.VT[3]], 
                                        [self.U[4], torch.diag(self.S[4]), self.VT[4]]],
                                       [self.Net.res2.layer1.bias, self.Net.res2.layer2.bias])
        
    
    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32 *32))
        x = self.activation(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.Net.fc2(x.T)
        return F.log_softmax(x, dim=1)
````

````python
def check_rank(rank):
    def our_func(x):
        return jnp.sum(jnp.square(jnp.sin(W2 @ jnp.cos(W1 @ x))))

    key = jax.random.PRNGKey(713)

    W1 = jax.random.uniform(key, (rank, 100), maxval=100., minval=0.,dtype=jnp.float32)
    W2 = jax.random.uniform(key, (100, rank), maxval=100., minval=0.,dtype=jnp.float32)

    x_samples = jax.random.uniform(key, (100, 100), maxval=1e9, minval=-1e9, dtype=jnp.float32)
    grad_f = jax.jit(jax.jacfwd(our_func))

    grads = grad_f(x_samples)
    grads = grads / jnp.sqrt(100)
    _, S, _ = jnp.linalg.svd(grads)
    pot_ranks = jnp.array([S[S > 10].shape[0], S[S > 1].shape[0], S[S>0.1].shape[0]])
    return pot_ranks

ranks = jnp.arange(1, 96, 1)
pot_ranks = np.zeros((ranks.shape[0], 3))
for i in ranks:
    pot_ranks[i-1] = (check_rank(i) == i)

pot_ranks = pot_ranks.T
cands = [10, 1, 0.1]
for it, rk in zip(pot_ranks, cands):
    print("Value of singular value to clip: {:.1f}, Accuracy of prediction: {:.3f}".format(rk, np.sum(it[it==1]) / it.shape[0]))
````
